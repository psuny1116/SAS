# 숫자 포맷

포맷 명령어|포맷 설명|포맷 예시|입력값|출력값
-|-|-|-|-
숫자.|자리 수만큼 정수 숫자값을 표시|5.|324.5678|325
숫자.숫자|앞자리 숫자만큼 전체 길이를 표현하고, 뒷자리 숫자만큼 소수로 표현(만약 입력값 길이가 앞자리 숫자를 초과할 경우 소수점 이하 길이가 축소될 수 있음)|6.2|324.5678|324.57
BEST숫자.|'.'을 포함하여 숫자 자리수만큼 수를 표현|BEST5.|324.5678|324.6
COMMA숫자.숫자|3자리 숫자값마다 ','를 표시|COMMA5.|3245.678|3,246
PERCENT숫자.숫자|주어진 숫자를 %로 표현함(단, 앞자리 숫자는 전체 길이, 뒷자리 숫자는 소수부분을 나타내는데 앞자리는 3보다 커야 함. %가 3의 위치만큼 차지함)|PERCENT8.2|0.3245678|32.46%
Z숫자.숫자|앞자리 수만큼 전체 숫자값을 표현하고 빈 길이만큼 0을 표시|Z7.2|324.5678|0324.57

```SAS
DATA TEST;
INPUT NAME $ AGE;
CARDS;
라이올라 324.5678
;
RUN;

PROC PRINT DATA=TEST;
VAR AGE;
FORMAT AGE 5.;
FORMAT AGE 6.2;
FORMAT AGE BEST5.;
FORMAT AGE COMMA5.;
FORMAT AGE PERCENT8.2;
FORMAT AGE Z7.2;
RUN;
```

- 위처럼 FORMAT을 한꺼번에 여러개 적으면 파이썬에서 print없이 적으면 가장 마지막 코드만 출력됐던 것처럼 FORMAT도 가장 마지막 포맷만 적용되어서 출력됨

# 문자 포맷

```SAS
DATA TEST;
INPUT NAME $ AGE;
CARDS;
라이올라 324.5678
;
RUN;

PROC PRINT DATA=TEST;
VAR NAME;
FORMAT NAME $12.;     /* NAME 문자 변수를 12자리까지 표현 */
RUN;
```

# 날짜 포맷

- SAS에서는 1960년 1월 1일을 숫자0으로 표시

포맷 명령어|포맷 설명|포맷 예시|입력값|출력값
-|-|-|-|-
DATE숫자.|일월년|DATE9.|3|4JAN1960
YYMMDD숫자.|연도-월-일|YYMMDD10.|3|1960-01-04
WEEKDATE.|요일, 월일, 연도|WEEKDATE.|3|Monday, January 4, 1960
WORDDATE.|월열, 연도|WORDDATE.|3|January 4, 1960
NLDATE숫자.|연도 월 일|NLDATE20.|3|1960년 01월 04일
YYMMN숫자.|연도 월|YYMMN6.|3|196001
MONYY숫자.|월 연도|MONYY7.|3|JAN1960
YEAR숫자.|연도|YEAR4.|3|1960
DATETIME숫자.숫자|일월년:시:분:초|DATETIME20.|3|01JAN1960:00:00:03
TIME숫자.숫자|시:분:초|TIME8.|3333|0:55:03
HHMM숫자.숫자|시:분|HHMM5.|3333|0:56
HOUR숫자.숫자|시각|HOUR5.|3333|1

```SAS
DATA TEST;
INPUT NAME $ NUMBER;
CARDS;
날짜 3
;
RUN;

PROC PRINT DATA=TEST;
VAR NUMBER;
FORMAT NUMBER DATE10.;
FORMAT NUMBER YYMMDD10.;
FORMAT NUMBER WEEKDATE.;
FORMAT NUMBER WORDDATE.;
FORMAT NUMBER NLDATE20.;
FORMAT NUMBER YYMMN6.;
FORMAT NUMBER MONYY7.;
FORMAT NUMBER YEAR4.;
FORMAT NUMBER DATETIME20.;
FORMAT NUMBER TIME8.;
FORMAT NUMBER HHMM5.;
FORMAT NUMBER HOUR5.;
RUN;
```

- 위처럼 FORMAT을 한꺼번에 여러개 적으면 파이썬에서 print없이 적으면 가장 마지막 코드만 출력됐던 것처럼 FORMAT도 가장 마지막 포맷만 적용되어서 출력됨
